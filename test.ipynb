```python
import os
import cv2
from facedb import FaceDB

# Function to get list of image files in a directory
def get_image_files(directory):
    image_files = []
    for file in os.listdir(directory):
        if file.endswith(('.jpg', '.jpeg', '.png')):
            image_files.append(os.path.join(directory, file))
    return image_files

# Create a FaceDB instance
db = FaceDB(
    path="/home/chanduchowdary/final/d_faces",
)

# Folder containing images
image_folder = "/home/chanduchowdary/final/d_faces/"

# Folder to save processed images
output_folder = "/home/chanduchowdary/final/processed_faces/"

# Get list of image files
image_files = get_image_files(image_folder)

# Iterate over each image file
for image_file in image_files:
    # Recognize faces in the image
    results = db.recognize(img=image_file)
    
    # Check if any faces were recognized
    if results:
        for result in results:
            face_id = result["id"]
            distance = result["distance"]
            
            # Create subdirectory for the recognized face if it doesn't exist
            face_folder = os.path.join(output_folder, face_id)
            os.makedirs(face_folder, exist_ok=True)
            
            # Save the recognized face only if distance is less than 0.5
            if distance < 0.5:
                # Load the original image
                img = cv2.imread(image_file)
                
                # Save the recognized face to the output folder with the image name
                output_filename = os.path.join(face_folder, f"{os.path.basename(image_file)}")
                cv2.imwrite(output_filename, img)
                print(f"Face recognized as '{face_id}' saved to '{output_filename}'")
            else:
                print(f"Face recognized as '{face_id}' has a distance of {distance}. Skipping...")
    else:
        print(f"No faces recognized in '{image_file}'")
```
